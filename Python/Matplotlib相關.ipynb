# 誤差折線圖

## 可調參數

xerr：x軸水平的誤差。ecolor：指定error bar的顏色。elinewidth：指定error bar的線條寬度

```python
x = np.arange(10)
y = 2.5 * np.sin(x / 20 * np.pi)
yerr = np.linspace(0.05, 0.2, 10) #y軸水平的誤差
plt.errorbar(x, y + 3, yerr=yerr, label='both limits')
```

# Histogram直方圖

## 可調參數

bins : 統計的區間分佈

range : 顯示的區間，在沒有給出bins時生效

density : 預設為false，True則顯示頻率統計結果

注意:頻率統計結果=區間數目/(總數*區間寬度)，和normed效果一致，官方推薦使用density

histtype : 可選{'bar', 'barstacked', 'step', 'stepfilled'}之一，預設為bar；step使用的是梯狀；stepfilled則會對梯狀內部進行填充，效果類似bar

align : 可選{'left', 'mid', 'right'}之一，預設為'mid'，控制柱狀圖的水平分佈，left或right會有部分空白區域。

```python
x=np.random.randint(0,100,100) 
bins=np.arange(0,101,10) #設定0~100的邊界值，間隔為10 
plt.hist(x,bins,color='fuchsia',alpha=0.5)#alpha為透明度，0是完全透明 
plt.xlabel('scores') 
plt.ylabel('count') 
plt.xlim(0,100); #設定x軸分布範圍
```

# 均勻子圖

## 可調參數

sharex和 sharey 分別表示是否共享橫軸和縱軸刻度

tight_layout函數可以調整子圖的相對大小

```python
fig, axs = plt.subplots(2, 5, figsize=(10, 4), sharex=True, sharey=True)
for i in range(2):
    for j in range(5):
        axs[i][j].scatter(np.random.randn(10), np.random.randn(10))
        axs[i][j].set_title('%d-%d'%(i+1,j+1))
        axs[i][j].set_xlim(-5,5)
        axs[i][j].set_ylim(-5,5)
        if i==1: 
            axs[i][j].set_xlabel('X axis')
        if j==0: 
            axs[i][j].set_ylabel('Y axis')
fig.tight_layout()
```

```python
plt.figure()
plt.subplot(2,2,1) #(總行數，總列數，當前子圖的index)
plt.plot([1,2], 'r')
plt.subplot(2,2,2)
plt.plot([1,2], 'b')
plt.subplot(2,2,4)   
plt.plot([1,2], 'g')
```

```python
r = 2 * np.random.rand(150)
theta = 2 * np.pi * np.random.rand(150)
area = 200 * r**2
colors = theta
plt.subplot(projection='polar')
plt.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=0.75) #cmap是colormap
```

# 非均勻子圖

## 可調參數

利用add_gridspec可以指定相對寬度比例 width_ratios 和相對高度比例height_ratios

spec = fig.add_gridspec(nrows=2, ncols=5, width_ratios=[1,2,3,4,5], height_ratios=[1,3]

```python
fig = plt.figure(figsize=(10, 4))
spec = fig.add_gridspec(nrows=2, ncols=5, width_ratios=[1,2,3,4,5], height_ratios=[1,3])
for i in range(2):
    for j in range(5):
        ax = fig.add_subplot(spec[i, j])
        ax.scatter(np.random.randn(10), np.random.randn(10))
        ax.set_title('%d-%d'%(i+1,j+1))
        if i==1: ax.set_xlabel('X axis')
        if j==0: ax.set_ylabel('Y axis')
fig.tight_layout()
```
