```python
from mlxtend.frequent_patterns import apriori 
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
df1 = pd.read_csv('C:/Users/jason/OneDrive/桌面/SCU/大數據行銷/data.csv', encoding="ISO-8859-1") #讀取原始資料
df1.head()
```

```python
df1 = df1[df1.Country == 'France'] #僅挑選法國資料進行後續分析
df1
```

```python
df1['Description'] = df1['Description'].str.strip() #將可能的空白移除
df1
```

```python
df1 = df1[df1.Quantity >0] #僅挑選商品數量大於0的記錄
df1
```

```python
basket = pd.pivot_table(data=df1,index='InvoiceNo',columns='Description',values='Quantity', 
                        aggfunc='sum',fill_value=0)

#繪製樞紐分析表，第三個參數為使用欄位、第四個參數為數值、第五個參數為數值表達方式，預設為mean平均數，這邊設定為sum總和。

basket.head()
```

```python
def convert_into_binary(x): 
    if x > 0:
        return 1
    else:
        return 0
```

```python
basket_sets = basket.applymap(convert_into_binary) #將basket資料表對應到上述自訂函式予以掃描
basket_sets.head(20)
```

```python
basket_sets.drop(columns=['POSTAGE'],inplace=True) #移除POSTAGE
basket_sets.head(40)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)

#使用apriori()產生頻繁資料集
#第一個參數為資料來源(basket_sets)、第二個參數為最小支持度
#第三個參數為是否顯示資料集內容

frequent_itemsets
```

```python
rules_mlxtend = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

#使用association_rules()產生關聯法則
#第一個參數為導入頻繁資料集(frequent_itemsets)、第二個參數為篩選機制設定，第三個參數為設定lift門檻值。

rules_mlxtend.head() 
```

```python
rules_mlxtend[ (rules_mlxtend['lift'] >= 4) & (rules_mlxtend['confidence'] >= 0.8) ]

#從所產生的rules_mlxtend關聯法則中提取出lift大於等於4且信賴度大於等於0.8的法則
```
