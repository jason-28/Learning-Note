```python
train1=train[['id','price','host_response_rate','host_acceptance_rate','neighbourhood_cleansed','accommodates','property_type','room_type','beds','bedrooms']]
test1=test[['id','neighbourhood_cleansed','accommodates','beds','bedrooms']]
train1=train1.dropna()
train1.info()
#test1=test1.dropna()
test1=test1.fillna(0)
test1.info()
```

```python
test1.describe()
```

```python
ml=linear_model.LinearRegression()
ml.fit(train1_x,train1_y)
```

```python
test1_y_prd = ml.predict(test1_x)
res=pd.DataFrame(test1_y_prd)
res['id'] = test['id']
res = res.rename(columns = {0:'price'}) 
res.head()
```
